### define a function to merge the given arrays and calculate the median named twoMedian

```python
# your code here
def selection_sort(arr):
 for i in range(len(arr)):
    index = i
    for j in range(i + 1, len(arr)):
        if arr[j] < arr[index]:
            index = j


    arr[i], arr[index] = arr[index], arr[i]

def twoMedian(x,y):

  arr=x+y
  selection_sort(arr)
  if len(arr)%2==0:
    median=(arr[len(arr)//2-1]+arr[len(arr)//2])/2
  else:
    median=arr[int(round(len(arr)/2))+1]
  return f"Median is {median}"
twoMedian([1,3,5],[2,5,7])
```
